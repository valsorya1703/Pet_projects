В столбце `salary_bin` больше всего значений с неуказанной категорией зарплаты. Попробуем дополнить часть, используя данные о диапазоне зарплат. Верхняя граница диапазона будет определять категорию. В случае, если в ячейке будет пропуск, воспользуемся нижней границей, дополнительно домжножив ее на медианное значение ширины диапазона.

```python
# вычисляем медиану ширины диапазона
kf_range = (data['salary_to']/data['salary_from']).median()
kf_range
```

```python
#Заполним возможные категории
data.loc[(data['salary_bin'] == 'ЗП не указана') & (pd.isna(data['salary_from']) == False), 'salary_bin'] = data['salary_from'].apply(lambda x: 'Меньше 100 тысяч' if x *
                                                                                                                                      kf_range < 100000 else ('От 100 тысяч до 200 тысяч' if x*kf_range <= 200000 else('От 200 тысяч до 300 тысяч' if x*kf_range <= 200000 else 'Больше 300 тысяч')))
```

Проверим, сколько категорий удалось восстановить.

```python
data['salary_bin'].value_counts()
```

Было восстановлено 233 категории.

### Исследовательский анализ данных

#### Средняя зарплата

Добавим новый столбец со средней зарплатой вакансии.

```python
def avg_sal(data):
    if pd.isna(data['salary_to'])==False and pd.isna(data['salary_from'])==False:
        return (data['salary_to']+data['salary_from'])/2
    elif pd.isna(data['salary_from']) == False:
        return data['salary_from']
    
```

```python
data['avg_salary']=data.apply(avg_sal, axis=1)
```

Сравним уровни зарплат между специальностями.

```python
(sns.boxplot(data = data, y = 'avg_salary', x='source', showfliers=False, showmeans=True)
    .set(title = 'З/п в зависимости от специальности', 
         xlabel = 'Специальность', ylabel = 'Уровень з/п'))
plt.show()
```

Медиана зарплат у DA находится около 100 тысяч, в то время как у DS она равна около 190 тысяч.

Делаем вывод, что hard-skills встречаются в 17,5 раз чаще.
